```python
import numpy as np

def objective_function(x):
    return 0.5 * x**2

def gradient(x):
    return x

def gradient_descent(gradient_fn, initial_x, learning_rate, num_iterations):
    trajectory = [initial_x]
    x = initial_x
    for _ in range(num_iterations):
        gradient = gradient_fn(x)
        x = x - learning_rate * gradient
        trajectory.append(x)
    return trajectory

def polyak_heavy_ball(gradient_fn, initial_x, learning_rate, momentum, num_iterations):
    trajectory = [initial_x]
    x = initial_x
    v = 0
    v_vector = []
    for _ in range(num_iterations):
        gradient = gradient_fn(x)
        v = momentum * v - learning_rate * gradient
        x = x + v
        v_vector.append(v)
        trajectory.append(x)
    return trajectory,v_vector

initial_x = 10
learning_rate = 0.1
momentum = 0.1
num_iterations = 200

gd_trajectory = gradient_descent(gradient, initial_x, learning_rate, num_iterations)
pb_trajectory,vvector = polyak_heavy_ball(gradient, initial_x, learning_rate, momentum, num_iterations)


print("Gradient Descent final values:", gd_trajectory)
print("Polyak's Heavy Ball final values:", pb_trajectory)
print("Polyak's velocity vector:", vvector)
```

A fundamental property of the (negative) gradient direction is that it always points perpendicular the contours of a function. This statement is universally true - and holds for any function and at all of its inputs.

```python
import numpy as np
import matplotlib.pyplot as plt

def objective_function(x):
    return x**2

def gradient(x):
    return 2 * x

def gradient_descent(gradient_fn, initial_x, learning_rate, num_iterations):
    trajectory = [initial_x]
    x = initial_x
    for _ in range(num_iterations):
        gradient = gradient_fn(x)
        x = x - learning_rate * gradient
        trajectory.append(x)
    return trajectory


initial_x = 10
learning_rate = 1.1
num_iterations = 20


gd_trajectory = gradient_descent(gradient, initial_x, learning_rate, num_iterations)

x_values = np.linspace(-initial_x * 1.2, initial_x * 1.2, 100)
y_values = objective_function(x_values)

plt.plot(x_values, y_values, label="Objective Function")
plt.plot(gd_trajectory, [objective_function(x) for x in gd_trajectory], marker='o', label="Gradient Descent", color='red')
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("Zig-Zag Problem of Gradient Descent")
plt.legend()
plt.show()
```
