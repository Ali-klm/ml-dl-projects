```python
import numpy as np
import matplotlib.pyplot as plt

class Perceptron:
    def __init__(self, learning_rate=0.01, epochs=100):
        self.weights = None
        self.learning_rate = learning_rate
        self.epochs = epochs

    def activation(self, x):
        return 1 if x >= 0 else 0

    def predict(self, inputs):
        summation = np.dot(inputs, self.weights[1:]) + self.weights[0]
        return self.activation(summation)

    def train(self, training_inputs, labels):
        self.weights = np.zeros(1 + training_inputs.shape[1])

        for _ in range(self.epochs):
            for inputs, label in zip(training_inputs, labels):
                prediction = self.predict(inputs)
                self.weights[1:] += self.learning_rate * (label - prediction) * inputs
                self.weights[0] += self.learning_rate * (label - prediction)

np.random.seed(0)
num_samples = 100
class_0 = np.random.normal(loc=0, scale=1, size=(num_samples, 2))
class_1 = np.random.normal(loc=2, scale=1, size=(num_samples, 2))

X = np.vstack([class_0, class_1])
y = np.hstack([np.zeros(num_samples), np.ones(num_samples)])

perceptron = Perceptron()
perceptron.train(X, y)

plt.figure(figsize=(8, 6))
plt.scatter(class_0[:, 0], class_0[:, 1], color='blue', label='Class 0')
plt.scatter(class_1[:, 0], class_1[:, 1], color='red', label='Class 1')

w0, w1, w2 = perceptron.weights
x_values = np.linspace(-3, 5, 100)
y_values = -(w1/w2) * x_values - (w0/w2)
plt.plot(x_values, y_values, color='green', label='Decision Boundary')

plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Perceptron Decision Boundary')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

class Perceptron:
    def __init__(self, learning_rate=0.01, epochs=100):
        self.weights = None
        self.learning_rate = learning_rate
        self.epochs = epochs

    def activation(self, x):
        return 1 if x >= 0 else 0

    def predict(self, inputs):
        summation = np.dot(inputs, self.weights[1:]) + self.weights[0]
        return self.activation(summation)

    def train(self, training_inputs, labels):
        self.weights = np.zeros(1 + training_inputs.shape[1])

        for _ in range(self.epochs):
            for inputs, label in zip(training_inputs, labels):
                prediction = self.predict(inputs)
                self.weights[1:] += self.learning_rate * (label - prediction) * inputs
                self.weights[0] += self.learning_rate * (label - prediction)

np.random.seed(0)
num_samples = 100

class_0_x = np.random.uniform(low=-2, high=0, size=num_samples)
class_0_y = np.random.uniform(low=-2, high=0, size=num_samples)
class_0 = np.vstack([class_0_x, class_0_y]).T

class_1_x = np.random.uniform(low=0, high=2, size=num_samples)
class_1_y = np.random.uniform(low=0, high=2, size=num_samples)
class_1 = np.vstack([class_1_x, class_1_y]).T

X = np.vstack([class_0, class_1])
y = np.hstack([np.zeros(num_samples), np.ones(num_samples)])

perceptron = Perceptron()
perceptron.train(X, y)

plt.figure(figsize=(8, 6))
plt.scatter(class_0[:, 0], class_0[:, 1], color='blue', label='Class 0')
plt.scatter(class_1[:, 0], class_1[:, 1], color='red', label='Class 1')

w0, w1, w2 = perceptron.weights
x_values = np.linspace(-2, 2, 100)
y_values = -(w1/w2) * x_values - (w0/w2)
plt.plot(x_values, y_values, color='green', label='Decision Boundary')

plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Perceptron Decision Boundary')
plt.legend()
plt.grid(True)
plt.show()
```
